```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "/home/shared/armeet/ultrasound/metrics/unet-3d-1_01-10-2024_22:38:02/metrics.pkl"
with open(path, 'rb') as f:
    data = np.array(pickle.load(f))
```

```python
data[0,:][23]
```

```python
train_metrics, test_metrics = np.array(data)
train_losses, train_accs = train_metrics[:,0], train_metrics[:, 1]
test_losses, test_accs = test_metrics[:,0], test_metrics[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(12, 6), layout="tight")

model_name = path.split("/")[-2]
x = range(0, len(train_losses))

fig.suptitle(f"{model_name}")

ax[0].set_title("L1 Loss")
ax[0].set_xlabel("epochs")
ax[0].scatter(x, train_losses, s=4, label="train")
ax[0].scatter(x, test_losses, s=4, label="test")
ax[0].legend()

ax[1].set_title("L1 Accuracy")
ax[1].set_xlabel("epochs")
ax[1].scatter(x, train_accs, s=4, label="train")
ax[1].scatter(x, test_accs, s=4, label="test")
ax[1].set_ylim(0, 1)
ax[1].legend()
fig.show()

final_train_loss, final_test_loss = train_losses[-1], test_losses[-1]
final_train_acc, final_test_acc = train_accs[-1], test_accs[-1]

print("final performance:")
print(f"train_loss={final_train_loss}, train_acc={final_train_acc}")
print(f"test_loss={final_test_loss}, train_acc={final_test_acc}")
```
